```python
from ucimlrepo import fetch_ucirepo
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")


data = fetch_ucirepo(id=891)

X = data.data.features
y = data.data.targets

df = X.copy()
df["Diabetes_Binary"] = y
```

```python
missing_counts = df.isnull().sum()

missing_df = pd.DataFrame({
    "feature": missing_counts.index,
    "missing_count": missing_counts.values,
    "missing_percent": (missing_counts.values / len(df)) * 100
})

missing_df = missing_df.sort_values("missing_percent", ascending=False)

missing_df
```

```python
# diabetes counts 
plt.figure(figsize=(6,4))

df["Diabetes_Binary"].value_counts(normalize=True).plot(
    kind="bar",
)

plt.title("Proportion of Diabetes Classes")
plt.ylabel("Proportion")

plt.xticks([0,1], ["No Diabetes", "Prediabetes/ Diabetes"], rotation = 0 )

print(df["Diabetes_Binary"].value_counts())
plt.show()
```

```python
# diabetes and sex

plt.figure(figsize=(6,4))

sns.barplot( x="Sex", y="Diabetes_Binary", data=df)
plt.xticks([0,1], ["Female", "Male"])

plt.title("Diabetes Risk by Sex")
plt.ylabel("Probability of Diabetes")

plt.show()
```

```python
#diabetes and incme
plt.figure(figsize=(8,4))

sns.barplot(x="Income", y="Diabetes_Binary", data=df)

plt.title("Diabetes Risk by Income Level")
plt.ylabel("Probability of Diabetes")

plt.show()

```

```python
#diabetes and education
plt.figure(figsize=(8,4))

sns.barplot(x="Education", y="Diabetes_Binary", data=df)

plt.title("Diabetes Risk by Education Level")
plt.ylabel("Probability of Diabetes")

plt.show()
```
